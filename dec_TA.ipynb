```python
df['time'] = pd.to_datetime(df['timestamp'])
df.set_index('timestamp', inplace=True)
df.head()
```

```python
df_pivot = df.pivot(index='time', columns='param_name', values='data')
df_pivot
```

```python
selected_columns = ['Reactive Power', 'Active Power', 'Apparent Power', 'PF']
df_filtered = df_pivot[selected_columns]

df_filtered.head()
```

```python
selected_columns = ['Reactive Power', 'Active Power', 'Apparent Power', 'PF']
df_filtered = df_pivot[selected_columns]

# Interpolate and then backward fill null values in all columns
df_filtered['Reactive Power'] = df_filtered['Reactive Power'].interpolate().bfill()
df_filtered['Active Power'] = df_filtered['Active Power'].interpolate().bfill()
df_filtered['Apparent Power'] = df_filtered['Apparent Power'].interpolate().bfill()

# Calculate 'fv'
fv = df_filtered['Active Power'] / df_filtered['Apparent Power']

# Fill null values in 'PF' column with 'fv' values
df_filtered['PF'] = df_filtered['PF'].fillna(fv)

# Verify the changes
print(df_filtered.isnull().sum())
```
